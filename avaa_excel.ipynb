<h2>Excel-datan avaaminen pandas dataframeen</h2>

Pandas kirjaston <b>read_excel</b> avaa Excel-muotoisen datan dataframeen.

Pelkällä tiedostonimellä avaat koodin kanssa samaan kansioon tallennetun Excel-datan, jossa data alkaa ensimmäisen taulukkovälilehden ensimmäisestä solusta.

Jos tiedosto on muussa kansiossa tai netissä, niin lisää tiedostopolku tai nettiosoite (esim. <i>'C://Users/Aki/Documents/data1.xlsx'</i>).

Datan alkuosaa voit avaamisen jälkeen katsoa komennolla <b>df.head()</b> ja loppuosaa komennolla <b>df.tail()</b>.

Jos data ei ala ensimmäisen taulukkovälilehden ensimmäisestä solusta (huomaat tämän <b>df.head()</b>-komennolla), niin tarvitset lisäparametreja <b>sheet_name</b>, <b>usecols</b>, <b>skiprows</b>. Esimerkiksi seuraava avaisi datan Sheet2-taulukkovälilehden sarakkeista B ja C hypäten ensimmäisen rivin yli: <i>df = pd.read_excel('data1.xlsx', sheet_name='Sheet2', usecols='B, C', skiprows=1)</i>

Hankalammissa tapauksissa kannattaa harkita datan siivoamista Excelissä ennen read_excel-komennon käyttämistä.
Lisätietoa read_excel käytöstä:
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_excel.html
