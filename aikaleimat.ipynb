<h2>Aikatiedon lukeminen merkkijonosta</h2>

Merkkijonona (tekstinä) esitetyt aikatiedot pitää muuntaa Pythonin ymmärtämiksi
aikatiedoiksi. Tämän voin tehdä esimerkiksi seuraavilla tavoilla:

<b>Tapa 1:</b>
<ul><li>Avaan datan ja katson, missä muodossa aikatiedot ovat?</li>
    <li>Muunnan aikatiedot pandas-kirjaston <b>to_datetime</b>-funktiolla ja sijoitan ne indeksiin.</li></ul>

<b>Tapa 2:</b>
<ul><li>Muunnan aikatiedot <b>strptime</b>-funktiolla.</li>
    <li>Sijoitan aikatiedot jo avaamisen yhteydessä indeksiin.</li></ul>

Muuntamisessa tarvitsen muotoilukoodeja, jotka löydän esimerkiksi osoitteesta:
https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior

```python
import pandas as pd
import numpy as np
```

Tapa 1

```python
# Avaan aikasarjatietoa sisältävän tiedoston
# ja katson aikatiedon esitysmuodon
df1 = pd.read_excel('http://taanila.fi/CO2.xlsx')
df1.head()
```

```python
# Sijoitan aikatiedon indeksiin
# to_datetime-funktio muuntaa merkkijonot aikatiedoiksi 
df1.index = pd.to_datetime(df1['Kuukausi'], format='%Y-%m')
df1 = df1.drop('Kuukausi', axis=1)
df1.head()
```

Tapa 2

```python
# Funktio, joka muuntaa merkkijonon aikatiedoiksi
def parser(x):
    return pd.datetime.strptime(x, '%Y-%m')

# Datan avaamisen yhteydessä sijoitan aikatiedot indeksiin
df2 = pd.read_excel('CO2.xlsx', index_col=0, date_parser=parser)
df2.head()
```

Tapa 1

```python
# Avaan aikasarjatietoa sisältävän tiedoston
# ja katson aikatiedon esitysmuodon
df3 = pd.read_csv('http://taanila.fi/Electric_Production.csv')
df3.head()
```

```python
# Funktio, joka muuntaa merkkijonon aikatiedoiksi
def parser(x):
    return pd.datetime.strptime(x,'%m/%d/%Y')

# Datan avaamisen yhteydessä sijoitan aikatiedot indeksiin
df4 = pd.read_csv('http://taanila.fi/Electric_Production.csv',
                 index_col=0, date_parser=parser)
df4.head()
```

<h2>Aikatietojen luominen</h2>

Voin luoda sarjan aikatietoja pandas-kirjaston <b>date_range</b>-funktiolla. Funktiolle annan parametreina täsmälleen kolme seuraavista: start, end, periods, freq. Lue lisää:

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.date_range.html

<b>freq</b>-parametrin mahdolliset arvot löydät seuraavasta:

https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#offset-aliases

```python
# Aikasarjan luvut
series = [500,350,250,400,450,350,200,300,350,200,150,400,550,
        350,250,550,550,400,350,600,750,500,400,650,850]

# Aikatietojen luominen vuosineljänneksittäin (Q)
index = pd.date_range(start='2013-03-31', periods=len(series), freq='Q')

# Dataframen luominen
df5 = pd.DataFrame(series, index=index)

# Otsikko aikasarjan luvuille
df5.columns=['Demand']
df5
```
